# Prototype Flight Map

Starter notebook for the `teamx_proto` interactive flight-track map. Run the setup cell first; it checks whether the required packages are available and makes the local prototype package importable.


```python
from pathlib import Path
import importlib.util
import sys

NOTEBOOK_DIR = Path.cwd()
if not (NOTEBOOK_DIR / "teamx_proto").exists():
    # Allows the notebook to run when Jupyter was started from the thesis_project root.
    NOTEBOOK_DIR = Path.cwd() / "TEAMx_prototype"

if str(NOTEBOOK_DIR) not in sys.path:
    sys.path.insert(0, str(NOTEBOOK_DIR))

REQUIRED_PACKAGES = [
    "numpy",
    "pandas",
    "xarray",
    "matplotlib",
    "panel",
    "holoviews",
    "hvplot",
    "bokeh",
    "geoviews"
]
missing = [pkg for pkg in REQUIRED_PACKAGES if importlib.util.find_spec(pkg) is None]
if missing:
    raise ImportError(
        "Missing packages: "
        + ", ".join(missing)
        + ". Install them with: pip install -r TEAMx_prototype/requirements.txt"
    )

print(f"Using prototype package from: {NOTEBOOK_DIR}")
```

```python
from teamx_proto import FlightMapApp
from teamx_proto.utils import create_large_dummy_dataset

ds = create_large_dummy_dataset()
ds
```

```python
app = FlightMapApp(
    ds,
    lat="lat",
    lon="lon",
    profile_dim="nprofile",
    time="time",
    thinning_step=20,
    hover_columns=["altitude"],
)
app.show()
```
